```python
import geopandas as gpd

# Load the GeoJSON file as a GeoDataFrame
gdf = gpd.read_file('stations.geojson')

# Display the first few rows
gdf.head()
```

```python
# Transform the x_offset column as specified
def transform_x_offset(val):
    if pd.isnull(val):
        return val
    try:
        v = int(val)
        if v < 0:
            return -1
        elif v > 0:
            return 1
        else:
            return 0
    except (ValueError, TypeError):
        return val  # leave non-integer values unchanged

import pandas as pd

gdf['x_offset'] = gdf['x_offset'].apply(transform_x_offset)
gdf['y_offset'] = gdf['y_offset'].apply(transform_x_offset)

gdf.to_file("stations_updated.geojson", driver="GeoJSON")

gdf.head()
```
